La **Programación Orientada a Objetos** organiza el código en torno a **objetos** que combinan datos y comportamiento. En este notebook aprenderás:

* El método `__init__` y el parámetro `self`.
* Atributos y métodos.
* El método `__str__`.
* Herencia.

## 1. Clases y objetos

Una **clase** es la plantilla; un **objeto** es una instancia de esa plantilla:

```python
class Perro:
    def __init__(self, nombre, raza):
        self.nombre = nombre   # atributo
        self.raza = raza

    def ladrar(self):          # método
        print(f"{self.nombre} dice: ¡Guau!")

mi_perro = Perro("Rex", "Labrador")   # crear objeto
mi_perro.ladrar()   # llamar método
print(mi_perro.nombre)   # acceder atributo
```

````python
# Ejemplo – clase Perro
class Perro:
    def __init__(self, nombre, raza):
        self.nombre = nombre
        self.raza = raza

    def ladrar(self):
        print(f"{self.nombre} dice: ¡Guau!")

    def __str__(self):
        return f"Perro({self.nombre}, {self.raza})"

rex = Perro("Rex", "Labrador")
luna = Perro("Luna", "Beagle")
rex.ladrar()
luna.ladrar()
print(rex)
````

## 2. Herencia

La **herencia** permite crear una clase nueva que reutiliza los atributos y métodos de otra:

```python
class Animal:
    def __init__(self, nombre):
        self.nombre = nombre

    def hablar(self):
        print("...")

class Gato(Animal):       # Gato hereda de Animal
    def hablar(self):     # sobreescribe el método
        print(f"{self.nombre} dice: ¡Miau!")

class Pato(Animal):
    def hablar(self):
        print(f"{self.nombre} dice: ¡Cuac!")

animales = [Gato("Misi"), Pato("Donald")]
for a in animales:
    a.hablar()
```

````python
# Ejemplo – herencia
class Animal:
    def __init__(self, nombre):
        self.nombre = nombre

    def hablar(self):
        print("...")

class Gato(Animal):
    def hablar(self):
        print(f"{self.nombre} dice: ¡Miau!")

class Pato(Animal):
    def hablar(self):
        print(f"{self.nombre} dice: ¡Cuac!")

animales = [Gato("Misi"), Pato("Donald"), Gato("Salem")]
for a in animales:
    a.hablar()
````
