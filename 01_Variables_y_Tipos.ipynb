# 01 – Variables, Tipos de Datos y Operadores

En este notebook aprenderás los conceptos más básicos de Python:

* **Variables** – almacenan valores en memoria.
* **Tipos de datos** – `int`, `float`, `str`, `bool`.
* **Operadores** – aritméticos, comparación y lógicos.
* **Entrada / Salida** – `input()` y `print()`.

## 1. Variables y tipos de datos

Una variable es un nombre que apunta a un valor. Python determina el tipo automáticamente:

```python
edad = 17          # int   – número entero
altura = 1.75      # float – número decimal
nombre = "Ana"     # str   – cadena de texto
aprobado = True    # bool  – verdadero / falso
```

Puedes comprobar el tipo con `type()`:

```python
print(type(edad))   # <class 'int'>
```

````python
# Ejemplo – ejecuta esta celda para ver los tipos
edad = 17
altura = 1.75
nombre = "Ana"
aprobado = True

print(type(edad), type(altura), type(nombre), type(aprobado))
````

## 2. Operadores aritméticos

| Operador | Significado       | Ejemplo      |
|----------|-------------------|--------------|
| `+`      | Suma              | `3 + 2 = 5`  |
| `-`      | Resta             | `5 - 1 = 4`  |
| `*`      | Multiplicación    | `4 * 3 = 12` |
| `/`      | División real     | `7 / 2 = 3.5`|
| `//`     | División entera   | `7 // 2 = 3` |
| `%`      | Módulo (resto)    | `7 % 2 = 1`  |
| `**`     | Potencia          | `2 ** 8 = 256`|

````python
# Ejemplo – operadores aritméticos
a = 15
b = 4
print("Suma:", a + b)
print("División real:", a / b)
print("División entera:", a // b)
print("Módulo:", a % b)
print("Potencia:", a ** b)
````

## 3. Operadores de comparación y lógicos

Los operadores de comparación devuelven `True` o `False`:

```python
==   !=   <   >   <=   >=
```

Los operadores lógicos combinan condiciones:

```python
and   or   not
```

Ejemplo:

```python
x = 7
print(x > 5 and x < 10)   # True
print(x == 3 or x == 7)   # True
print(not x > 5)           # False
```
