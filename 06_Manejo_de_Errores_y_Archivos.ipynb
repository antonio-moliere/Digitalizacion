## 1. ¿Qué es una excepción?

Cuando ocurre un error en tiempo de ejecución Python lanza una **excepción** que detiene el programa. Ejemplos comunes:

| Excepción            | Causa habitual                              |
|----------------------|---------------------------------------------|
| `ZeroDivisionError`  | Dividir entre cero                          |
| `ValueError`         | Conversión incorrecta (`int("abc")`)         |
| `IndexError`         | Índice fuera de rango en lista              |
| `KeyError`           | Clave no existe en diccionario              |
| `FileNotFoundError`  | El archivo no existe                        |
| `TypeError`          | Operación sobre tipos incompatibles         |

## 2. try / except

```python
try:
    numero = int(input("Introduce un número: "))
    resultado = 100 / numero
    print("Resultado:", resultado)
except ValueError:
    print("Error: debes introducir un número entero.")
except ZeroDivisionError:
    print("Error: no se puede dividir entre cero.")
else:
    print("Todo fue bien.")       # solo si no hubo excepción
finally:
    print("Fin del programa.")    # siempre se ejecuta
```

````python
# Ejemplo – captura de excepciones
def dividir_seguro(a, b):
    try:
        resultado = a / b
    except ZeroDivisionError:
        return "Error: división entre cero"
    else:
        return resultado

print(dividir_seguro(10, 2))
print(dividir_seguro(10, 0))
````

## 3. Lanzar excepciones con `raise`

Puedes lanzar tus propias excepciones para señalar condiciones de error:

```python
def raiz_cuadrada(n):
    if n < 0:
        raise ValueError(f"No se puede calcular la raíz de {n}")
    return n ** 0.5

try:
    print(raiz_cuadrada(-4))
except ValueError as e:
    print("Error:", e)
```

## 4. Lectura y escritura de archivos

```python
# Escribir un archivo
with open("notas.txt", "w", encoding="utf-8") as f:
    f.write("Ana: 9\n")
    f.write("Luis: 6\n")

# Leer el archivo completo
with open("notas.txt", "r", encoding="utf-8") as f:
    contenido = f.read()
    print(contenido)

# Leer línea por línea
with open("notas.txt", "r", encoding="utf-8") as f:
    for linea in f:
        print(linea.strip())
```

El uso de `with` garantiza que el archivo se cierra aunque ocurra un error.

Modos de apertura: `"r"` (leer), `"w"` (escribir, borra si existe), `"a"` (añadir al final).

````python
# Ejemplo – escribir y leer un archivo
with open("/tmp/ejemplo.txt", "w", encoding="utf-8") as f:
    for i in range(1, 6):
        f.write(f"Línea {i}\n")

with open("/tmp/ejemplo.txt", "r", encoding="utf-8") as f:
    for linea in f:
        print(linea.strip())
````

## 5. Trabajar con JSON

JSON es un formato muy usado para guardar y compartir datos estructurados:

```python
import json

datos = {"nombre": "Ana", "nota": 9, "asignaturas": ["Mates", "Inglés"]}

# Guardar en archivo JSON
with open("/tmp/alumno.json", "w", encoding="utf-8") as f:
    json.dump(datos, f, ensure_ascii=False, indent=2)

# Leer desde archivo JSON
with open("/tmp/alumno.json", "r", encoding="utf-8") as f:
    cargado = json.load(f)

print(cargado["nombre"])   # Ana
```
