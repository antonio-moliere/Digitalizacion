# 03 – Funciones

Las **funciones** permiten agrupar código que se va a reutilizar, darle un nombre y llamarlo cuando sea necesario. En este notebook aprenderás:

* Definir y llamar funciones (`def`).
* Parámetros, argumentos y valores por defecto.
* La instrucción `return`.
* Alcance de variables (scope).
* Funciones que llaman a otras funciones.
* Introducción a la recursividad.

## 1. Definir y llamar una función

```python
def saludar():
    print("¡Hola!")

saludar()   # llamada a la función
```

Una función puede recibir **parámetros** y devolver un resultado con `return`:

```python
def sumar(a, b):
    return a + b

resultado = sumar(3, 5)
print(resultado)   # 8
```

````python
# Ejemplo – función básica con parámetros y return
def saludar(nombre):
    return f"¡Hola, {nombre}!"

print(saludar("Ana"))
print(saludar("Carlos"))
````

## 2. Parámetros con valor por defecto

Puedes asignar un valor por defecto a un parámetro. Si no se pasa ese argumento, se usa el valor por defecto:

```python
def presentar(nombre, saludo="Hola"):
    print(f"{saludo}, {nombre}!")

presentar("Ana")              # Hola, Ana!
presentar("Carlos", "Buenos días")  # Buenos días, Carlos!
```

## 3. Funciones que trabajan con listas

```python

    total = 0

        total += n
    return total
```

## 4. Alcance de variables (scope)

Las variables creadas **dentro** de una función son locales y no existen fuera de ella:

```python
def mi_funcion():
    x = 10   # variable local
    print(x)

mi_funcion()  # imprime 10
print(x)      # ¡Error! x no existe aquí
```

Las variables **globales** se pueden leer dentro de funciones, pero para modificarlas hay que declarar `global`.

## 5. Recursividad

Una función **recursiva** se llama a sí misma. Necesita un **caso base** para detenerse:

```python
def factorial(n):
    if n == 0:          # caso base
        return 1
    return n * factorial(n - 1)   # llamada recursiva

print(factorial(5))   # 120  (5*4*3*2*1)
```

````python
# Ejemplo – factorial recursivo
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)

for i in range(8):
    print(f"{i}! = {factorial(i)}")
````
